```python
# import time time sleep(1)
import requests
from bs4 import BeautifulSoup
# 컬럼 생성
# 페이지에 있는 컬럼: 22

columns = ['고유 번호', '선수명', '포지션','연도', '팀명', 'AVG', 'G', 'PA', 'AB', 'R', 'H', '2B', '3B', 'HR', 'TB', 'RBI', 'SB', 'CS', 'BB', 'HBP','SO','GDP','SLG','OBP','E']
# 선수들의 기록을 크롤링할 URL 템플릿
base_url = "https://www.koreabaseball.com/Futures/Player/HitterTotal.aspx?playerId={}"
# 최종 데이터를 저장할 리스트 생성
result_list = []
# 크롤링할 선수들의 player_num 리스트
for player_num in range(80000, 90000):
    # player_num을 포함한 완전한 URL 생성
    url = base_url.format(player_num)
    # 해당 URL에 GET 요청
    response = requests.get(url)
    if response.status_code == 200:
        # BeautifulSoup 객체 생성
        soup = BeautifulSoup(response.content, "html.parser")
        # 데이터를 저장할 리스트 생성
        data_list = []
        # 기록 데이터
        # <tbody> 태그 찾기 기록데이터
        tbody_tag = soup.find('tbody')
        # <tr> 태그들 찾기 , 기록 데이터
        tr_tags = tbody_tag.find_all('tr')
        # 이름 데이터
        # <span> 태그 찾기
        name_tag = soup.find('span', {'id': 'cphContents_cphContents_cphContents_ucPlayerProfile_lblName'})
        # 포지션 데이터
        # <span> 태그 찾기
        position_tag = soup.find('span', {'id': 'cphContents_cphContents_cphContents_ucPlayerProfile_lblPosition'})
        # 데이터가 있는 경우에만 이름 추출
        if name_tag:
            name = name_tag.get_text()
            # 데이터가 있는 경우에만 이름 추출
            if position_tag:
                position = position_tag.get_text()
        # 연도 데이터 추출
            for tr_tag in tr_tags:
                td_tags = tr_tag.find_all('td')
                data_row = [td.get_text() for td in td_tags]
                # “기록이 없습니다.” 데이터를 제외하고 리스트에 추가
                if "기록이 없습니다." not in data_row:
                    data_list.append(data_row)
            # 데이터가 존재하는 경우 결과 저장
            if data_list:
                for data_row in data_list:
                    data_row.insert(0, player_num)
                    data_row.insert(1, name)
                    data_row.insert(2, position)
                    print(data_row)
                    
    else:
        continue
# 데이터가 존재하는 경우 결과 저장
    if data_list:
        result_list.extend(data_list)
# 딕셔너리 데이터를 DataFrame으로 변환
df = pd.DataFrame(result_list, columns=columns)
df
```
